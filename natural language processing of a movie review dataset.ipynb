```python


import nltk
from nltk.corpus import movie_reviews
movie_reviews.categories()
from nltk import pos_tag
nltk.download('wordnet')

```

```python
len(movie_reviews.fileids())
```

```python
movie_reviews.words(movie_reviews.fileids()[5])
```

```python
documents=[]
for cate in movie_reviews.categories() :
    for fileid in movie_reviews.fileids(cate):
        documents.append((movie_reviews.words(fileid),cate))
documents[0:1500]
```

```python
import random 
random.shuffle(documents)
documents[0:5]
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
from nltk.corpus import stopwords
import string
stop=set(stopwords.words('english'))
punctuations=list(string.punctuation)
stop.update(punctuations)
```

```python
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    else: 
        return wordnet.NOUN

```

```python
def clean_reviews(words):
    output_words=[]
    for w in words:
        if w.lower() not in stop:
            pos=pos_tag([w])
            clean_words=lemmatizer.lemmatize(w,pos=get_simple_pos(pos[0][1]))
            output_words.append(clean_words.lower())
    return output_words
            
    
```

```python
documents=[(clean_reviews(document),cate) for document,cate in documents]
```

```python
training_documents=documents[0:1500]
testing_documents=documents[1500:]
```

```python
all_words=[]
for doc in training_documents:
    all_words+=doc[0]
    
```

```python
import nltk
freq=nltk.FreqDist(all_words)
```

```python
common=freq.most_common(3000)
```

```python
features=[i[0] for i in common]
features
```

```python
def get_feature_dict(words):
    curr_feat={}
    word_set=set(words)
    for w in features:
        curr_feat[w]=w in word_set
    return curr_feat

    
    
```

```python
training_data=[(get_feature_dict(doc),cat) for doc,cat in training_documents]
testing_data=[(get_feature_dict(doc),cat) for doc,cat in training_documents]

training_data[0]
```

```python
from nltk import NaiveBayesClassifier
```

```python
classifier= NaiveBayesClassifier.train(training_data)
nltk.classify.accuracy(classifier,testing_data)
```

```python
from sklearn.svm import SVC
from nltk.classify.scikitlearn import SklearnClassifier
svc=SVC()
classifier_sklearn=SklearnClassifier(svc)
classifier_sklearn.train(training_data)
```

```python
nltk.classify.accuracy(classifier_sklearn,testing_data)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
classifier_sklearn1=SklearnClassifier(rf)
classifier_sklearn1.train(training_data)
nltk.classify.accuracy(classifier_sklearn1,testing_data)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
categories=[categories for document,categories in documents ]
text_documents=[" ".join(document) for document,category in documents]
text_documents
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(text_documents,categories,train_size=0.8,test_size=0.2)
```

```python
count_vec=CountVectorizer(max_features=3000,ngram_range=(1,2))
```

```python
x_transform=count_vec.fit_transform(x_train)
x_transform.todense()
```

```python
count_vec.get_feature_names()
x_test_features=count_vec.transform(x_test)
x_test_features
```

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(x_transform,y_train)
svc.score(x_test_features,y_test)
```
